### Sorting and Ordering data 

```python
import os
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold
os.chdir('C:\\Users\\haihe\\Desktop\\Classes\\Intro to Python\\week7labs\\data')
os.getcwd()
gh_teams = pd.read_csv("github_teams.csv")
```

```python
# Select `human-bot` teams that have a `bot_members_count` value greater than and equal to 2.

gh_teams[(gh_teams.Team_type == 'human-bot') & (gh_teams.bot_members_count >= 2)]
```

```python
# Find the `human` teams that are `Large` and have a `human_gini` value greater than and equal to 0.75.

gh_teams[(gh_teams.Team_type == 'human') & (gh_teams.Team_size_class == 'Large')&(gh_teams.human_gini <= 0.75)]
```

```python
# How many teams are in the `Small` or `Large` category?

small_and_Large = gh_teams[gh_teams.Team_size_class.isin(['Small', 'Large'])]
value_counts = small_and_Large['Team_size_class'].value_counts()
print(value_counts)
```

```python
# How many teams are in the `Small` or `Large` cateogry with a `human_gini` value less than and equal to 0.20?

size_and_gini = gh_teams[gh_teams.Team_size_class.isin(['Small', 'Large'])&(gh_teams.human_gini <= 0.20)]
value_counts2 = size_and_gini['Team_size_class'].value_counts()
print(value_counts2)
```

```python
# How many `human-bot` teams are in the `Medium` category?

humanbot = gh_teams[gh_teams.Team_size_class.isin(['Medium'])&(gh_teams.Team_type == 'human-bot')]
value_counts3 = humanbot['Team_size_class'].value_counts()
print(value_counts3)
```

```python
# Create a subsample of 50% of your data.

teams_subsample = gh_teams.sample(frac = 0.5, replace = False)  
len(teams_subsample.index)
```

```python
# Create samples for a 8-fold cross validation test.

kf = KFold(n_splits = 8)  

for train, test in kf.split(gh_teams):
    print("%s %s" % (train, test))

gh_teams.iloc[train]
gh_teams.iloc[test]
```

```python
# Select columns that are numeric and save it as a new DataFrame.

new_teams = gh_teams.select_dtypes(include=[np.number]).copy()
```

```python
# Remove the columns `bot_PRReviewComment` and `bot_MergedPR` from the DataFrame.

new_teams.drop(['bot_PRReviewComment', 'bot_MergedPR'], axis=1, inplace=True)
```

```python
# Save the columns `Team_size_class` and `human_members_count` as a new DataFrame.

new_teams1 = gh_teams[['Team_size_class','human_members_count']].copy()
print(new_teams1)
```
